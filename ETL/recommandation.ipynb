```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import requests
import numpy as np
```

```python
import pandas as pd
df = pd.DataFrame(requests.get("http://localhost:3000/api/manga").json())
df
```

```python
mangaId = df[['id','description']]
mangaId['after_transfo'] = np.nan
mangaId['description'].fillna("no description",inplace = True)
for i in range(len(mangaId)):
        text = mangaId.iloc[i,1]
        tokens = word_tokenize(text)
        lemmatizer = WordNetLemmatizer()
        token = [lemmatizer.lemmatize(token) for token in tokens]
        stop_words = stopwords.words('english')
        token = [token for token in tokens if token.lower() not in stop_words]
        mangaId.iloc[i,2]= " ".join(token)

        

        

vectorizer = TfidfVectorizer()
vector = vectorizer.fit_transform(mangaId['after_transfo'])
similarity = cosine_similarity(vector)

similarity


    
```

```python

mangaId['Recommandation']  = np.empty((len(mangaId), 0)).tolist()
for i in range(len(mangaId)):
    listOfIndex = np.argsort(similarity[i])[-11:][:10][::-1]
    listOfId = []
    for j in range (len(listOfIndex)):
        listOfId.append(mangaId['id'].iloc[listOfIndex[j]])
    mangaId['Recommandation'].iloc[i]= listOfId


mangaId
```
