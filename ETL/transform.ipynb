# Transformation de la BDD

```python
df["Favorites"] = df["Favorites"].apply(lambda x: int(x.replace(',','')))
df["Members"] = df["Members"].apply(lambda x: int(x.replace(',','')))

df['Chapters'] = pd.to_numeric(df['Chapters'], errors='coerce').fillna(0).astype('int64')
df['Volumes'] = pd.to_numeric(df['Volumes'], errors='coerce').fillna(0).astype('int64')

```

```python
# Séparation des dates en deux colonnes 
def split_date_s(i):
    try:
        
        return i.split('to')[0]
    except:
        return i 
def split_date_e(i):
    try:
        return i.split('to')[1]
    except:
        return '00 00,0000'
    
def parse_date(date_str):
    try:
        
        return datetime.strptime(date_str.strip(), '%b %d, %Y').date()
    except ValueError:
        try:
           
            return datetime.strptime(date_str.strip(), '%b %Y').date()
        except ValueError:
            try:
               
                return datetime.strptime(date_str.strip(), '%Y').date()
            except ValueError:
                return datetime(1,1,1).date()
    
df['Start_Date']= df['Published'].apply(split_date_s)
df['End_Date'] = df['Published'].apply(split_date_e)
df['Start_Date'] = df['Start_Date'].apply(parse_date)
df['End_Date'] = df['End_Date'].apply(parse_date)
```

# Faisons notre algorithme pour la recommmandation qui sera basé sur du NLP

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import requests
import numpy as np
```

```python
import pandas as pd
df = pd.DataFrame(requests.get("http://localhost:3000/api/manga").json())
df
```

```python
mangaId = df[['id','description']]
mangaId['after_transfo'] = np.nan
mangaId['description'].fillna("no description",inplace = True)
for i in range(len(mangaId)):
        text = mangaId.iloc[i,1]
        tokens = word_tokenize(text)
        lemmatizer = WordNetLemmatizer()
        token = [lemmatizer.lemmatize(token) for token in tokens]
        stop_words = stopwords.words('english')
        token = [token for token in tokens if token.lower() not in stop_words]
        mangaId.iloc[i,2]= " ".join(token)

        

        

vectorizer = TfidfVectorizer()
vector = vectorizer.fit_transform(mangaId['after_transfo'])
similarity = cosine_similarity(vector)

similarity
```

```python
Manga['recommandation']  = np.empty((len(mangaId), 0)).tolist()
for i in range(len(mangaId)):
    listOfIndex = np.argsort(similarity[i])[-11:][:10][::-1]
    listOfId = []
    for j in range (len(listOfIndex)):
        listOfId.append(mangaId['id'].iloc[listOfIndex[j]])
    Manga['recommandation'].iloc[i]= listOfId


Manga
```
